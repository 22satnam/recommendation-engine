```python
import pandas as pd
import numpy as np
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2015'
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
df = df1.append(df2.append(df3.append(df4, ignore_index=True), ignore_index=True), ignore_index=True)
```

```python
df
```

```python
dfb_2015 = df[['Title', 'Director', 'Cast', 'Genre']]
```

```python
dfb_2015
max = 0
for i in df.Cast:
    l = len(i.split(','))
    if max < l:
      max = l
print(max)
actor_no = []
for i in range(1,max+1):
    actor_no.append('Actor'+str(i))
print(actor_no)
```

```python
df[actor_no] = df.Cast.str.split(",",expand=True)
```

```python
df
```

```python
droping_actor_no=[]
for i in range(4,max+1):
  droping_actor_no.append('Actor'+str(i))
print(droping_actor_no)
```

```python
dfb_2015 = df.drop(droping_actor_no, axis = 1)
```

```python
dfb_2015 = dfb_2015.rename(columns={'Title':'movie_title', 'Director':'director_name','Actor1':'actor_1_name','Actor2':'actor_2_name','Actor3':'actor_3_name','Cast':'comb','Genre':'genres'})
```

```python
dfb_2015
```

```python
new_dfb15 = dfb_2015.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
new_dfb15
```

```python
new_dfb15['actor_2_name'] = new_dfb15['actor_2_name'].replace(np.nan, 'unknown')
new_dfb15['actor_3_name'] = new_dfb15['actor_3_name'].replace(np.nan, 'unknown')
```

```python
new_dfb15['movie_title'] = new_dfb15['movie_title'].str.lower()
```

```python
new_dfb15['comb'] = new_dfb15['actor_1_name'] + ' ' + new_dfb15['actor_2_name'] + ' '+ new_dfb15['actor_3_name'] + ' '+ new_dfb15['director_name'] +' ' + new_dfb15['genres']
```

```python
new_dfb15
```

```python
old_df = pd.read_csv('2014.csv')
```

```python
old_df
```

```python
final_df = old_df.append(new_dfb15,ignore_index=True)
```

```python
final_df
```

```python
final_df.isna().sum()
```

```python
final_df = final_df.dropna(how='any')
```
