```python
import pandas as pd
import numpy as np
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2013'
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
df = df1.append(df2.append(df3.append(df4, ignore_index=True), ignore_index=True), ignore_index=True)
```

```python
df
```

```python
dfb_2013 = df[['Title', 'Director', 'Cast', 'Genre']]
```

```python
dfb_2013[['Actor1','Actor2','Actor3','Actor4', 'Actor5', 'Actor6', 'Actor7', 'Actor8', 'Actor9', 'Actor10', 'Actor11', 'Actor12', 'Actor13']] = dfb_2013.Cast.apply(lambda x : pd.Series(str(x).split(", ")))
```

```python
dfb_2013
```

```python
for i in range(4, 14):
  x = 'Actor'
  dfb_2013.drop(x+str(i),axis=1,inplace=True)
```

```python
dfb_2013
```

```python
dfb_2013 = dfb_2013.rename(columns={'Title':'movie_title', 'Director':'director_name','Actor1':'actor_1_name','Actor2':'actor_2_name','Actor3':'actor_3_name','Cast':'comb','Genre':'genres'})
```

```python
dfb_2013
```

```python
new_dfb13 = dfb_2013.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
new_dfb13
```

```python
new_dfb13['actor_2_name'] = new_dfb13['actor_2_name'].replace(np.nan, 'unknown')
new_dfb13['actor_3_name'] = new_dfb13['actor_3_name'].replace(np.nan, 'unknown')
```

```python
new_dfb13['movie_title'] = new_dfb13['movie_title'].str.lower()
```

```python
new_dfb13['comb'] = new_dfb13['actor_1_name'] + ' ' + new_dfb13['actor_2_name'] + ' '+ new_dfb13['actor_3_name'] + ' '+ new_dfb13['director_name'] +' ' + new_dfb13['genres']
```

```python
new_dfb13
```

```python
old_df = pd.read_csv('2012.csv')
```

```python
old_df
```

```python
final_df = old_df.append(new_dfb13,ignore_index=True)
```

```python
final_df
```

```python
final_df.isna().sum()
```

```python
final_df = final_df.dropna(how='any')
```
